```python
from sklearn.model_selection import train_test_split
```

```python
usecols = ['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']

X = df_train[usecols]
y = df_train['Survived']
```

```python
DEFAULT_AGE = df_train['Age'].mean()
X['Age'] = X['Age'].fillna(DEFAULT_AGE)//10
```

```python
X
```

```python
DEFAULT_CABIN_CODE = 'X'
X['Cabin_code'] = X['Cabin'].fillna(DEFAULT_CABIN_CODE).apply(lambda x:x[0])
```

```python
DEFAULT_EMBARKED = 'S'
X['Embarked'] = X['Embarked'].fillna(DEFAULT_EMBARKED)
```

```python
onehot_cols = ['Sex','Embarked','Cabin_code']
num_cols = ['Pclass','Age','SibSp','Parch','Fare']
X_training = pd.concat([X[num_cols],pd.get_dummies(X[onehot_cols])],axis=1)
```

```python
X_train.shape
```

```python
X = X_training
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Define model
lr = LogisticRegression()
```

```python
# Train model
lr.fit(X_train,y_train)
```

```python
# Prediction
df_result = pd.DataFrame(y_train)
df_result['predicted'] = lr.predict(X_train)
```
